```python
import cv2
import mediapipe as mp
```

```python
cap = cv2.VideoCapture(0)
mpPose = mp.solutions.pose
pose = mpPose.Pose()
mpDraw = mp.solutions.drawing_utils

while True:
    succes, img = cap.read()
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result = pose.process(imgRGB) 
    
    if result.pose_landmarks:
        mpDraw.draw_landmarks(img, result.pose_landmarks, mpPose.POSE_CONNECTIONS) 
    
    cv2.imshow("Image", img)
    cv2.waitKey(1)
```

```python
cap = cv2.VideoCapture(0)
mpPose = mp.solutions.pose
pose = mpPose.Pose()
mpDraw = mp.solutions.drawing_utils

while True:
    succes, img = cap.read()
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result = pose.process(imgRGB) 
    
    if result.pose_landmarks:
        mpDraw.draw_landmarks(img, result.pose_landmarks, mpPose.POSE_CONNECTIONS) 
        
        for id, lm in enumerate(result.pose_landmarks.landmark): # Nos ayudara a identificar cada uno de los puntos (landmarks)
                # Print de coordenadas de cada uno de los landmarks
                print(id, lm) 
                
                # Print de coordenadas de cada uno de los puntos de la mano en pixeles
                h, w, c = img.shape
                centerX, centerY = int(lm.x*w), int(lm.y*h)
                print(id, centerX, centerY)
                
                # Podemos usar estas coordenadas como queramos
                # Estamos haciendo que el punto de la nariz sea morado y mas grande
                if id == 0:
                    cv2.circle(img, (centerX, centerY), 10, (255,0,255), cv2.FILLED)
    
    cv2.imshow("Image", img)
    cv2.waitKey(1)
```
